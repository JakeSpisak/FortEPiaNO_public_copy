```python
import matplotlib
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
import numpy as np
```

```python
N=20
yv=np.linspace(0.01,20,N)
plt.loglog(yv, [y*y/(np.exp(y/1.4)+1) for y in yv],'bo-')
yv=np.logspace(-2,np.log10(20),N)
plt.loglog(yv, [y*y/(np.exp(y/1.4)+1) for y in yv],'ro-')
plt.show()
plt.close()
```

```python
folder="output"
zdat=np.loadtxt("%s/z.dat"%folder)
d1dat=np.loadtxt("%s/nuDens_diag1.dat"%folder)
d2dat=np.loadtxt("%s/nuDens_diag2.dat"%folder)
d3dat=np.loadtxt("%s/nuDens_diag3.dat"%folder)
```

```python
for x in d3dat:
    plt.loglog(yv, [d/y**2 for y,d in zip(yv,x[1:])])
plt.show()
```
